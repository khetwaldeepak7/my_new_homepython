```python
# as we know grp represent ke  lia humko 2 axis chahiyeh,ek x-axis y-axis so to tke 2 value we will use numpy
import numpy as np
from matplotlib import pyplot as pl
x=np.arange(11)
print(x)
```

```python
y=2*x    # we got 2 second y axis values
print(y)
```

```python
pl.plot(x,y)  # in side the pl.plot is a method which we take x,y axis
pl.show()
```

```python
# to give label to the x and y axis is
```

```python
pl.xlabel("this is x axis")  # to  give x -axis name
pl.ylabel("this is for y axis")  # y -axis name
pl.plot(x,y,color='g',linestyle=':',linewidth=5)  # # to give color ,style and width to the line
pl.show()
```

# to plot 2 liness together in a graph

```python
import numpy as np
from matplotlib import pyplot as pl
x=np.arange(11)
y1=2*x    # 1 line chahiye
y2=3*x    # 2 line chahiyeh 
# to title to the Graph
pl.title("this is used for title")
pl.xlabel("this is x axis")  #--------->>>>>>>>>> to  give  x -axis name
pl.ylabel("this is for y axis")  #-------->>>>>>>>  to  give  y -axis name
pl.plot(x,y1,color='b',linestyle=':',linewidth=5)  # # ----- >>>>  to give color ,style and width to the line
pl.plot(x,y2,color='r',linestyle='-',linewidth=5) ###----->>>>>>
pl.grid(True)  # grid sey piche graph mai line aa jayenge,mean grid sey piche line aya
pl.show()      #  show is used to display 
```
